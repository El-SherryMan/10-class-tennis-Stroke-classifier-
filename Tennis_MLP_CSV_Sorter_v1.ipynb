```python
import os

def standardize_dataset_names(base_dir):
    for root, _, files in os.walk(base_dir):
        # Extract the name of the immediate parent directory (e.g., "AA000000")
        parent_folder_name = os.path.basename(root)
        
        for file in files:
            if file.endswith('.csv'):
                parts = file.split('_')
                
                # Replace the old ID (parts[1]) with the parent folder's name
                if len(parts) >= 2:
                    parts[1] = parent_folder_name
                    new_name = '_'.join(parts)
                    
                    old_path = os.path.join(root, file)
                    new_path = os.path.join(root, new_name)
                    
                    # Rename the file in-place
                    os.rename(old_path, new_path)

# Execute the script on the specified directory path
dataset_path = r"C:\Users\sheri\OneDrive\Desktop\tennisDS1"
standardize_dataset_names(dataset_path)
```
